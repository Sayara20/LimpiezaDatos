```python
#funcion para buscar las unidades de medidas
def get_first_capacity(text, word_to_find):
    # Manejo de entradas no válidas (NaN o no string)
    if not isinstance(text, str):
        return ''
    patron = rf"(\d+\s*{word_to_find})"
    # Usamos re.search() para encontrar la PRIMERA coincidencia.
    try:
        match = re.search(patron, text, re.IGNORECASE)
        if match:
            # Devuelve la coincidencia completa (ej. '11oz' o '12 oz')
            return match.group(1)
        else:
            # Si no hay coincidencia, devuelve una cadena vacía
            return ''         
    except re.error:
        # Manejo de errores de RegEx, aunque es poco probable con este patrón
        return ''

```

```python
#buscar capacidad oz
unidad_a_buscar = 'oz'
df[f'Capacidad_{unidad_a_buscar}'] = df['features'].apply(
    lambda x: get_first_capacity(x, unidad_a_buscar)
)
df.head()
```

```python
#quita acentos funciona
import unicodedata
def quitar_acentos(texto):
    
    nfkd_form = unicodedata.normalize('NFD', str(texto))
    sin_acentos = ''.join([c for c in nfkd_form if unicodedata.category(c) != 'Mn'])
    return sin_acentos

new_df.loc[:,'features']=new_df['features'].apply(quitar_acentos)
new_df.loc[:,'category']=new_df['category'].apply(quitar_acentos)
new_df.head()
```

```python
#Funcion para la impieza de categoria
def limpiar_columna(df, columna):
    df_limpio = df.copy()
    df_limpio[columna] = df_limpio[columna].astype(str)
    # Quitar acentos
    df_limpio[columna] = df_limpio[columna].apply(
        lambda x: ''.join(
            c for c in unicodedata.normalize('NFD', x)
            if unicodedata.category(c) != 'Mn'
        )
    )
    df_limpio[columna] = df_limpio[columna].str.replace(r'\d+', '', regex=True)#elimina numeros
    df_limpio[columna] = df_limpio[columna].str.replace('oz', '', regex=True)
    df_limpio[columna] = df_limpio[columna].str.translate(str.maketrans('', '', string.punctuation)) # Quitar signos de puntuación
    df_limpio[columna] = df_limpio[columna].str.replace(r'[^\w\s]', '', regex=True) # Quitar caracteres especiales
    df_limpio[columna] = df_limpio[columna].str.replace(r'\s+', ' ', regex=True).str.strip()# Quitar espacios dobles y espacios al inicio/final
    df_limpio[columna] = df_limpio[columna].str.lower() #  Convertir a minúsculas

    
    return df_limpio
```

```python
# Limpiar solo la columna 'categoria'
new_df= limpiar_columna(new_df, 'category')
new_df.head(10)
```
